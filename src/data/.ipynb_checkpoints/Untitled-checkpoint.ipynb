```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('rtc_timeseries_data.csv')
```

```python
df.head()
```

```python
df = df.T.reset_index()
```

```python
dt_col = df['index'].iloc[1:]
```

```python
dt_col = pd.to_datetime(dt_col).dt.to_period("Q")
```

```python
dt_col = dt_col.to_list()
dt_col.insert(0, 'Date')
```

```python
df['index'] = dt_col
```

```python
df.head()
```

```python
df = df.T
```

```python
df.iloc[0] = pd.concat([pd.Series('State'), pd.to_datetime(df.iloc[0][1:])])
```

```python
df.to_csv('rtc_timeseries_data(modified).csv', index=False, header=False)
```
